```python
import os
import librosa
import numpy as np
import pandas as pd
import tensorflow_hub as hub
from tqdm import tqdm

# Function to load audio and preprocess for the model
def load_and_preprocess_audio(audio_path, duration=5, sr=32000):
    audio, _ = librosa.load(audio_path, sr=sr, duration=duration)
    if len(audio) < sr * duration:
        audio = np.pad(audio, (0, max(0, sr * duration - len(audio))), "constant")
    return audio

# Function to save embeddings to a file
def save_embeddings(embeddings, output_dir, filename):
    os.makedirs(output_dir, exist_ok=True)
    np.save(os.path.join(output_dir, filename + '_embeddings.npy'), embeddings)
```

```python
# Function to extract features using the pre-trained model
import tensorflow as tf
import tensorflow_hub as hub

# Load the pre-trained model from TensorFlow Hub
model_url = '/root/.cache/kagglehub/models/google/bird-vocalization-classifier/tensorFlow2/bird-vocalization-classifier/8'
model = hub.load(model_url)

# Corrected function to extract features using the pre-trained model
def extract_features(audio):
    # Prepare the input as expected by the model; this often involves specific tensor shapes and types
    audio_tensor = tf.constant(audio[np.newaxis, :], dtype=tf.float32)  # Ensure correct shape and data type

    # Use the appropriate signature for inference, usually 'serving_default'
    inference_func = model.signatures['serving_default']
    outputs = inference_func(audio_tensor)  # Using the serving default function

    # Extract and return the desired outputs
    logits = outputs['output_0'].numpy()  # Adjust 'output_0' based on the actual output key
    embeddings = outputs['output_1'].numpy()  # Adjust 'output_1' based on the actual output key
    return logits, embeddings


# Process a batch of audio files and extract embeddings
def process_audio_files(metadata, base_audio_dir, output_base_dir):
    results = []
    for index, row in tqdm(metadata.iterrows(), total=metadata.shape[0]):
        audio_path = os.path.join(base_audio_dir, row['filename'])
        if os.path.exists(audio_path):
            audio = load_and_preprocess_audio(audio_path)
            _, embeddings = extract_features(audio)
            save_embeddings(embeddings, output_base_dir, os.path.splitext(row['filename'])[0])
            results.append(f"Processed and saved embeddings for: {row['filename']}")
        else:
            results.append(f"Audio file not found: {row['filename']}")
    return results
```

```python

# Data Preparation
metadata_path = '/kaggle/input/birdclef-2024/train_metadata.csv'
metadata = pd.read_csv(metadata_path)
filtered_metadata = metadata[metadata['latitude'].between(8, 21) & metadata['longitude'].between(72, 78)]

# Specify directories
base_audio_dir = '/kaggle/input/birdclef-2024/train_audio'
output_base_dir = '/kaggle/working/processed_embeddings'

# Process all audio files
results = process_audio_files(filtered_metadata, base_audio_dir, output_base_dir)

# Print some of the results
for result in results[:10]:
    print(result)
```
